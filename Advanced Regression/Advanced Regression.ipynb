```python
df.drop(columns = dropcolumns, inplace=True)
```

```python
def fill(df, fill_columns, mode="mean"):
    dfcopy = df.copy()
    for column in fill_columns:
        if dfcopy[column].dtype == "O":
            dfcopy[column] = dfcopy[column].fillna("None")
        elif mode=="median":
            dfcopy[column] = dfcopy[column].fillna(dfcopy[column].median())
        elif mode=="mean":
            dfcopy[column] = dfcopy[column].fillna(dfcopy[column].mean())
    return dfcopy
```

```python
df = fill(df, fillcolumns)
```

```python
df_dummied = pd.get_dummies(df,drop_first=True)
```

```python
df_test= df_dummied[df_dummied["SalePrice"].isnull()==True]
df_train= df_dummied[df_dummied["SalePrice"].isnull()==False]
df_test.reset_index
df_train.reset_index
```

```python
y, X=df_train['SalePrice'], df_train.drop(columns=['SalePrice'], axis=1)
```

```python
from sklearn.linear_model import ElasticNet, Lasso, LinearRegression, Ridge, BayesianRidge
from sklearn.linear_model import OrthogonalMatchingPursuit, HuberRegressor
from sklearn.kernel_ridge import KernelRidge
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor, ExtraTreesRegressor, AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor 
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
en, l, lr, r, br, omp = ElasticNet(), Lasso(), LinearRegression(), Ridge(), BayesianRidge(), OrthogonalMatchingPursuit()
hr, kr, rf, ada= HuberRegressor(), KernelRidge(), RandomForestRegressor(), AdaBoostRegressor()
gb, dt= GradientBoostingRegressor(), DecisionTreeRegressor()
```

```python
algorithms = [en, l, lr, r, br, omp,hr, kr, rf, ada,gb, dt]
names = ["ElasticNet", "Lasso", "LinearR", "Ridge", "BayesianRidge", "OrthogonalMatchingPursuit",
        "Huber", "KernelRidge", "RandomForest", "AdaBoost",
        "GradientBoosting", "DecisionTree"]
r2score = []
mse = []
regressor = []
def regression(X,y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.2, random_state=101)
    for i, algo in enumerate(algorithms):
        algorithms[i].fit(X_train,y_train)
    for i, algo in enumerate(algorithms):
        r2score.append("%.2f" % (r2_score(y_test, algorithms[i].predict(X_test))*100))
        mse.append("%.2f" % (mean_squared_error(y_test,algorithms[i].predict(X_test)))**0.5)
        regressor.append(names[i])


    met= pd.DataFrame(columns= ["Regressor","r2_score", "mse"])
    met["Regressor"] = regressor
    met["r2_score"] = r2score
    met["mse"] = mse
    return met.sort_values(by="r2_score", ascending=False)
```

```python
regression(X,y)
```

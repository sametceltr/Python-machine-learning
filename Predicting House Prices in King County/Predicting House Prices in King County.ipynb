```python
from sklearn.linear_model import LinearRegression #science kit
from sklearn.model_selection import train_test_split
```

```python
df["age"]= 2022-df["yr_built"]
df["agen"]= 2022-df["yr_renovated"]
df["total"]= df["sqft_living"]+df["sqft_lot"]+df["sqft_basement"]+df["sqft_living15"]+df["sqft_lot15"]
df
```

```python
df["total"].describe()
```

```python
df= df[df["total"] < 125000]
df.shape
```

```python
df= df[df["age"] < 100]
df.shape
```

```python
df= df[df["bedrooms"] < 6]
df.shape
```

```python
df= df[df["bathrooms"] < 5]
df.shape
```

```python
df= df[df["sqft_living"] < 5000]
df.shape
```

```python
df= df[df["sqft_above"] < 4500]
df.shape
```

```python
df["bedrooms"]= df["bedrooms"]**2
```

```python
df["bathrooms"]= df["bathrooms"]**2
```

```python
#df["waterfront"]= df["waterfront"]*4
```

```python
x, y= df[["sqft_living","grade","sqft_above", "sqft_living15", "bathrooms",  "lat", "waterfront", "floors","age","zipcode"]], df[["price"]]
```

```python
x.head()
```

```python
x= pd.get_dummies(x, drop_first=True)
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state= 13)


lr= LinearRegression()
regmodel= lr.fit(x_train, y_train)
y_pretest= regmodel.predict(x_test)
y_pretest
```

```python
(mean_squared_error(y_test, y_pretest))**0.5
```

```python
r2_score(y_test, y_pretest)
```

```python
from sklearn.linear_model import Ridge, Lasso
```

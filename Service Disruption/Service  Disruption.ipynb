```python
y, X= df_dummied["fault_severity"], df_dummied.drop(columns=["fault_severity","id"], axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30, random_state=101)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score
```

```python
m= MultinomialNB()
k= KNeighborsClassifier()
log= LogisticRegression()
gbc= GradientBoostingClassifier()
r= RandomForestClassifier() 
d= DecisionTreeClassifier()
```

```python
algorithms=[m,k,log,gbc,r,d]
names=["MultinomialNB",'K Nearest','Logistic','GradientBoosting','RandomForest','DecisionTree'] 
def algo_test (A,b,algorithms=algorithms,names=names):
    # fit the data 
    for i in range(len(algorithms)):
        algorithms[i]=algorithms[i].fit(A,b)
    accuracy=[]
    precision=[]
    recall=[]
    f1=[]
    for i in range(len(algorithms)):
        accuracy.append(accuracy_score(y_test,algorithms[i].predict(X_test)))
        precision.append(precision_score(y_test,algorithms[i].predict(X_test), average=None).max())
        recall.append(recall_score(y_test,algorithms[i].predict(X_test), average=None).max())
        f1.append(f1_score(y_test,algorithms[i].predict(X_test), average=None).max())
    metrics=pd.DataFrame(columns=['Accuracy','Precision','Recall','F1 Score'],index=names)
    metrics['Accuracy']=accuracy 
    metrics['Precision']=precision 
    metrics['Recall']=recall 
    metrics['F1 Score']=f1
    #.sort_values('F1',ascending=False)  
    return metrics.sort_values(by="F1 Score", ascending=False)
```

```python
algo_test(X_train, y_train)
#Takes too much time to compile. Decision Tree has best scores.
```

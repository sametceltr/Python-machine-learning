```python
df2 = df_origin[["Order_Items.product_name", "Products.short_description"]].drop_duplicates()
df2
```

```python
func = lambda x: str(x)
df2["Order_Items.product_name"], df2["Products.short_description"] = df2["Order_Items.product_name"].apply(func),df2["Products.short_description"].apply(func)
```

```python
#df_origin[["Order_Items.product_name", "Products.short_description"]].drop_duplicates()
df2=df2.groupby('Order_Items.product_name', as_index=False).agg({'Order_Items.product_name' : 'first', 'Products.short_description' : ','.join})
```

```python
from textblob import TextBlob
import nltk
from nltk import word_tokenize
import string
import re
from textblob import Word
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

def data_cleaning(df):
    stop_words= list(stopwords.words("english"))
    
    
    df["text"] = df["Products.short_description"].apply(lambda x: re.sub(',', ' ', x))
    df["text"] = df["text"].apply(lambda x: re.sub(r'[^\w\s]', ' ', x))
    df["text"] = df["text"].apply(lambda x: word_tokenize(x.lower()))
    df["text"]= df["text"].apply(lambda x: [word for word in x if word not in stop_words])
    df["text"]= df["text"].apply(lambda x: " ".join(x))
    df["text"]= df["text"].apply(lambda x: WordNetLemmatizer().lemmatize(x))
    
    return df
```

```python
df3 = data_cleaning(df2)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()



corpus = df3["text"]
corpcolumns = df3["Order_Items.product_name"]
cv_t = cv.fit_transform(corpus)

pd.DataFrame(cv_t.toarray(),columns=cv.get_feature_names(),index=corpcolumns)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfid = TfidfVectorizer()

corpus = df3["text"]
corpcolumns = df3["Order_Items.product_name"]
tfd_t=tfid.fit_transform(corpus)
pd.DataFrame(tfd_t.toarray(),columns=tfid.get_feature_names(),index=corpcolumns)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns

df_cossim = pd.DataFrame(cosine_similarity(tfd_t))
df_cossim.columns = corpcolumns.tolist()
df_cossim.index = corpcolumns.tolist()
```
